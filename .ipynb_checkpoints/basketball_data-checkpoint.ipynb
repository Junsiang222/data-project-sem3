## modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
features = ['age', 'g', 'gs', 'mp', 'orb', 'drb', 'trb',
            'ast', 'stl', 'blk', 'tov', 'pf']


# df['pos'] = df['pos'].astype('category').cat.codes
# df['pos']
```

```python

X = df[features]
y = df['ppg']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
base_model = LinearRegression()
rf_model = RandomForestRegressor(random_state=42)
gb_model = GradientBoostingRegressor(random_state=42)
```

```python
base_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
```

```python
def evaluate_model(model,X_test,y_test,name='Model'):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test,y_pred)
    r2 = r2_score(y_test,y_pred)

    print(f"--{name}--")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R2: {r2:.2f}")
    print("-" * 20)
    return r2
```

```python
df.ppg.describe()
```

```python
r2_lr = evaluate_model(base_model, X_test, y_test, 'LinearRegression')
r2_rf = evaluate_model(rf_model, X_test, y_test, 'RandomForestRegression')
r2_gb = evaluate_model(gb_model, X_test, y_test, 'GradientBoostRegression')
```

```python
def predict_ppg(age, g, gs, mp, orb, drb, trb, ast, stl, blk, tov, pf):
    model = joblib.load('ppg_model.joblib')
    input_arr = np.array([[ 
        age, g, gs, mp, orb, drb, trb,
        ast, stl, blk, tov, pf
    ]])
    predict = model.predict(input_arr)
    return predict.round(2)[0]
```

```python
import gradio as gr
import numpy as np
import joblib

# ------------------------
# Prediction function
# ------------------------
def predict_ppg(age, g, gs, mp, orb, drb, trb, ast, stl, blk, tov, pf):
    model = joblib.load('ppg_model.joblib')  # Your trained model
    input_arr = np.array([[ 
        age, g, gs, mp, orb, drb, trb,
        ast, stl, blk, tov, pf
    ]])
    prediction = model.predict(input_arr)
    return round(float(prediction[0]), 2)

# ------------------------
# Gradio interface components
# ------------------------
with gr.Blocks() as demo:
    gr.Markdown("<h1 style='text-align:center; color:#FF4500;'>🏀 NBA PPG Predictor</h1>")
    gr.Markdown("Enter player stats to predict their average points per game (PPG)")

    with gr.Row():
        with gr.Column():
            age = gr.Number(label="Age", value=25)
            g = gr.Number(label="Games Played (G)", value=82)
            gs = gr.Number(label="Games Started (GS)", value=50)
            mp = gr.Number(label="Minutes per Game (MP)", value=30.0)
            orb = gr.Number(label="Offensive Rebounds per Game (ORB)", value=2.0)
        
        with gr.Column():
            drb = gr.Number(label="Defensive Rebounds per Game (DRB)", value=4.0)
            trb = gr.Number(label="Total Rebounds per Game (TRB)", value=6.0)
            ast = gr.Number(label="Assists per Game (AST)", value=5.0)
            stl = gr.Number(label="Steals per Game (STL)", value=1.0)
        
        with gr.Column():
            blk = gr.Number(label="Blocks per Game (BLK)", value=0.5)
            tov = gr.Number(label="Turnovers per Game (TOV)", value=2.0)
            pf = gr.Number(label="Fouls per Game (PF)", value=2.0)

    output = gr.Textbox(label="Predicted PPG", interactive=False)

    btn = gr.Button("Predict PPG")
    btn.click(
        predict_ppg, 
        inputs=[age, g, gs, mp, orb, drb, trb, ast, stl, blk, tov, pf],
        outputs=output
    )

# ------------------------
# Launch Gradio
# ------------------------
demo.launch()
```
